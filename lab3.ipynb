```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_digits()
data, labels = dataset['data'], dataset['target']
X_train, X_test, Y_train, Y_test = train_test_split(data, labels, random_state=0)
```

```python
scr = []
n_estimators = np.arange(1,30)
for i in n_estimators:
    clf = RandomForestClassifier(max_depth=10, random_state=0, n_estimators=i)
    clf.fit(X_train, Y_train)
    scr.append(clf.score(X_test,Y_test))
plt.plot(n_estimators, scr,'r')
plt.xlabel('Количество деревьев')
plt.ylabel('Точность')
plt.title('Метрика качества от количества деревьев')
plt.show()
```

```python
scr = []
n_depths = range(2, 30)
for i in n_depths:
    clf = RandomForestClassifier(max_depth = i, random_state=0)
    clf.fit(X_train, Y_train)
    scr.append(clf.score(X_test, Y_test))


plt.plot(n_depths, scr,'r')
plt.xlabel('Максимальная глубина')
plt.ylabel('Точность')
plt.title('Качество обучения от глубины деревьев')
plt.show()
```
